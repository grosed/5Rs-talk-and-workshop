```julia
#Tested using Julia 1.8.5
function approx1(z)
    return (1/2)*( 1 + (1 - exp(-2*z*z/pi))^(1/2))
end

function approx2(z)
    return 1 - (1/sqrt(2*pi))*((exp(-z*z/pi))/(z + 0.8*exp(-0.4*z)) )
end

function approx3(z)
    k = sqrt(2/pi)

    return (exp(2*k*z))/(1 + exp(2*k*z))
end

function approx4(z)
    t = 1/(1+0.33267*z)
    a1 = 0.4361836
    a2 = 0.1201676
    a3 = 0.937298

    return 1 - (a1*t - a2*t*t + a3*t*t*t)*(exp(-z*z/2)/sqrt(2*pi))
end

function approx5(z)
    a = (1 + sqrt(1 - 2*pi*pi + 6*pi))/(2*pi)
    b = 2*pi*a*a
    c = sqrt(1 + b*z*z)/(1 + a*z*z)

    P0 = sqrt(pi/2)

    return 1 - (exp(-z*z/2)/(sqrt(2*pi)*z))*(1 - (c)/(P0*z + sqrt( P0*P0*z*z + exp(-z*z/2)*c )))
end

function approx6(z)
    y = sqrt(2/pi)*z*(1 + 0.044715*z*z)
    return 0.5*(1 + tanh(y))
end

function approx7(z)
    y = 0.806*z*(1-0.018*z)

    return 1 - 0.5*( 1 - (1 - exp(-y*y))^0.5 )
end

function approx8(z)
    return 1 - 0.5*exp( -0.717*z - 0.416*z*z)
end

function approx9(z)
    if z <= 2.7
        return 1 - 0.5*exp(-(z*z + 1.2*z^0.8)/2)
    else
        return (1/(sqrt(2*pi)*z))*exp(-z*z/2)
    end
end

function approx10(z)
    y = 4.2*pi*(z/(9 - z))
    return 1 - 1/(1 + exp(y))
end

function approx11(z)
    return 1/( 1 + exp(-sqrt(pi)*( 0.9*z + 0.0418198*z*z*z - 0.0004406*z*z*z*z*z )))
end

function approx12(z)
    return (( (4 - pi)*z + sqrt(2*pi)*(pi - 2) )/( (4 - pi)*z*z*sqrt(2*pi) + 2*pi*z + 2*sqrt(2*pi)*(pi - 2) ))*exp(-z*z/2)
end

function approx13(z)
    a1 = 5.575192695
    a2 = 12.77436324
    b1 = 14.38718147
    b2 = 31.53531977

    return ((z*z + a1*z + a2)/(sqrt(2*pi)*z*z*z + b1*z*z + b2*z + 2*a2))*exp(-z*z/2)
end

function approx14(z)
    return 0.5 + 0.5*sqrt(1 - exp(-sqrt(pi/8)*z*z))
end

function approx15(z)
    ugh = -(z*z/2)*( 4/pi + 0.147*z*z/2 )/(1 + 0.147*z*z/2)
    return 0.5*( 1 + sqrt(1 - exp(ugh)))
end

function approx16(z)
    if z<=3.36
        H1 = tanh(-0.2695*z)
        H2 = tanh(0.5416*z)
        H3 = tanh(0.4134*z)

        return 0.5 - 1.136*H1 + 2.47*H2 - 3.013*H3
    else
        return 1.0
    end
end

function approx17(z)
    H11 = tanh(1.280022196 - 0.720528073*z)
    H12 = tanh(0.033142223 - 0.682842425*z)

    if z <= 3.6
        return 0.46375416 + 0.065687194*H11 - 0.602383931*H12
    else
        return 1.0
    end
end

function approx18(z)
    return 1 - (1/sqrt(2*pi))*(exp(-z*z/2)/( 0.226 + 0.64*z + 0.33*sqrt(z*z + 3) ))
end


function approx19(z)
    H1 = tanh(0.043 + 0.2624*z)
    H2 = tanh(-1.687 - 0.519*z)
    H3 = tanh(-1.654 + 0.5044*z)

    y = 0.125 + 3.611*H1 - 4.658*H2 + 4.982*H3

    return 1/(1 + exp(-y))
end

function approx20(z)
    w = 0.268

    return w*approx5(z) + (1 - w)*approx13(z)
end
6
function approx21(z)
    return 1 - exp(-z*z/2)/( 44/79 + 8*z/5 + (5/6)*sqrt(z*z + 3))
end 
```

```julia
function approxAll(z)
    return [approx1(z),approx2(z),approx3(z),approx4(z),approx5(z),approx6(z),approx7(z),approx8(z),approx9(z),approx10(z),approx11(z),approx12(z),approx13(z),approx14(z),approx15(z),approx16(z),approx17(z),approx18(z),approx19(z),approx20(z),approx21(z)]
end
```
